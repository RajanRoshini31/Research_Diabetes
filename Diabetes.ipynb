```python
# Medical diagnosis: Given the symptoms exhibited in a patient and a database of anonymized patients,
# a model of this decision problem could be used by a program to provide decision support to medical professionals.

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load data
data = pd.read_csv("/content/diabetes_prediction_dataset.csv")

# Preprocessing
X = data.drop('diabetes', axis=1)
X = pd.get_dummies(X, columns=['gender', 'smoking_history'])
y = data['diabetes']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model training
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Prediction
y_pred = clf.predict(X_test)

# Accuracy calculation
accuracy = accuracy_score(y_test, y_pred)

# Identify diabetic and non-diabetic patients
diabetic = data.iloc[X_train.index]
diabetic = diabetic[y_pred == 1]
non_diabetic = diabetic[y_pred == 0]
# Save the identified diabetic patients to a CSV file
diabetic.to_csv('diabetic-patients.csv', index=False)

# Print the accuracy of the model
print(f"Accuracy: {accuracy}")

# Confirmation message
print("Diabetic patients saved to 'diabetic-patients.csv'")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten, Dropout

# Load data
data = pd.read_csv("/content/diabetes_prediction_dataset.csv")

# Preprocessing
X = data.drop('diabetes', axis=1)
X = pd.get_dummies(X, columns=['gender', 'smoking_history'])
y = data['diabetes']

# Standardize the feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reshape the data for CNN (1D convolution expects 3D input: samples, time_steps, features)
X_reshaped = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.3, random_state=42)

# Build CNN model
model = Sequential([
    Conv1D(32, kernel_size=3, activation='relu', input_shape=(X_reshaped.shape[1], 1)),
    Dropout(0.3),
    Conv1D(64, kernel_size=3, activation='relu'),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=1)

# Prediction
y_pred_probs = model.predict(X_test)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()

# Accuracy calculation
accuracy = accuracy_score(y_test, y_pred)

# Identify diabetic and non-diabetic patients in the test set
X_test_original = pd.DataFrame(X_test.reshape(X_test.shape[0], -1), columns=X.columns)
X_test_original['diabetes_pred'] = y_pred
diabetic_patients = X_test_original[X_test_original['diabetes_pred'] == 1]
diabetic_patients.to_csv('diabetic-patients.csv', index=False)

# Print the accuracy of the model
print(f"Accuracy: {accuracy}")

# Confirmation message
print("Diabetic patients saved to 'diabetic-patients.csv'")
```

```python
# Import Required Modules.
from statistics import mean, stdev
from sklearn import preprocessing
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
import pandas as pd

# Load diabetes dataset.
data = pd.read_csv("/content/diabetes_prediction_dataset.csv")

# Preprocessing: Separate features (X) and target variable (y)
X = data.drop('diabetes', axis=1)
y = data['diabetes']

# Feature Scaling for input features.
scaler = preprocessing.MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Create classifier object.
lr = LogisticRegression(random_state=42)

# Create StratifiedKFold object (10-fold).
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# List to store accuracy results.
lst_accu_stratified = []

# Perform Stratified K-Fold Cross Validation
for train_index, test_index in skf.split(X_scaled, y):
    X_train_fold, X_test_fold = X_scaled[train_index], X_scaled[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]

    # Train the model
    lr.fit(X_train_fold, y_train_fold)

    # Evaluate the model
    lst_accu_stratified.append(lr.score(X_test_fold, y_test_fold))

# Print the output.
print('List of possible accuracy:', lst_accu_stratified)
print('\nMaximum Accuracy That can be obtained from this model is:', max(lst_accu_stratified) * 100, '%')
print('\nMinimum Accuracy:', min(lst_accu_stratified) * 100, '%')
print('\nOverall Accuracy:', mean(lst_accu_stratified) * 100, '%')
print('\nStandard Deviation is:', stdev(lst_accu_stratified))
```

```python
# Cross-Validation StratifiedKFold
# Import Required Modules.
from statistics import mean, stdev
from sklearn import preprocessing
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
import pandas as pd

# Load diabetes dataset.
data = pd.read_csv("/content/diabetes_prediction_dataset.csv")

# Preprocessing: Separate features (X) and target variable (y)
X = data.drop('diabetes', axis=1)
y = data['diabetes']

# Encode categorical variables using One-Hot Encoding
X = pd.get_dummies(X, columns=['gender', 'smoking_history'])

# Feature Scaling for input features.
scaler = preprocessing.MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Create classifier object.
lr = LogisticRegression(random_state=42)

# Create StratifiedKFold object (10-fold).
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# List to store accuracy results.
lst_accu_stratified = []

# Perform Stratified K-Fold Cross Validation
for train_index, test_index in skf.split(X_scaled, y):
    X_train_fold, X_test_fold = X_scaled[train_index], X_scaled[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]

    # Train the model
    lr.fit(X_train_fold, y_train_fold)

    # Evaluate the model
    lst_accu_stratified.append(lr.score(X_test_fold, y_test_fold))

# Print the output.
print('List of possible accuracy:', lst_accu_stratified)
print('\nMaximum Accuracy That can be obtained from this model is:', max(lst_accu_stratified) * 100, '%')
print('\nMinimum Accuracy:', min(lst_accu_stratified) * 100, '%')
print('\nOverall Accuracy:', mean(lst_accu_stratified) * 100, '%')
print('\nStandard Deviation is:', stdev(lst_accu_stratified))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Dropout

# Load the dataset
data = pd.read_csv("/content/diabetes_prediction_dataset.csv")

# Preprocess the data
X = data.drop('diabetes', axis=1)
X = pd.get_dummies(X, columns=['gender', 'smoking_history'])  # One-hot encoding
y = data['diabetes']

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reshape the data for RNN (samples, timesteps, features)
# Here, we treat each feature as a timestep. So, we reshape the data to 3D (samples, 1, features)
X_reshaped = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.3, random_state=42)

# Build the RNN model
model = Sequential([
    SimpleRNN(64, activation='relu', input_shape=(X_reshaped.shape[1], 1), return_sequences=True),
    Dropout(0.2),
    SimpleRNN(32, activation='relu', return_sequences=False),
    Dropout(0.2),
    Dense(1, activation='sigmoid')  # Binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=1)

# Prediction
y_pred_probs = model.predict(X_test)
y_pred = (y_pred_probs > 0.5).astype(int)

# Accuracy calculation
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print(f"RNN Model Accuracy: {accuracy * 100:.2f}%")
```
